```python
#@title Code to Import Packages
import os
import pandas as pd
import plotly.express as px
```

```python
#@title Code to plot the graph
poverty           = pd.read_csv("ddf--datapoints--extreme_poverty_percent_people_below_190_a_day--by--geo--time.csv")
poverty_per_year  = poverty.where(poverty["time"]>=1997).groupby(['time']).agg({'extreme_poverty_percent_people_below_190_a_day':'mean'}).reset_index()
fig = px.line(poverty_per_year, x="time", y="extreme_poverty_percent_people_below_190_a_day",
              width = 1000,
              height=400)

fig.update_layout(title=go.layout.Title(text="Extreme Poverty last 20 years"),
    xaxis=go.layout.XAxis(title=go.layout.xaxis.Title(text="Years")),
    yaxis=go.layout.YAxis(title=go.layout.yaxis.Title(text="Percentage of Extreme Poverty")))

fig.show()
```

> **We can see that the value for extreme povery was 17.28 in 1997 but in  2017 is 3.53, which indicates that extreme poverty has more than halved.**

# Problem 3
### Use visualization to investigate the relationship between income (GDP per capita), life expectancy, and child mortality over time.

```python
#@title Code to create the plotly plots

life_expent   = 'ddf--datapoints--life_expectancy_years--by--geo--time.csv'
child_morta   = 'ddf--datapoints--child_mortality_0_5_year_olds_more_years_version_7--by--geo--time.csv'
child_morta_2 = 'ddf--datapoints--child_mortality_0_5_year_olds_dying_per_1000_born--by--geo--time.csv'
geo_country   = 'ddf--entities--geo--country.csv'



gdp         = pd.read_csv("ddf--datapoints--income_per_person_long_series--by--geo--time.csv")
life        = pd.read_csv(life_expent)
chld_mor    = pd.read_csv(child_morta_2)
geoMap      = pd.read_csv(geo_country)[['country','world_4region']]
gdp_var     = 'income_per_person_long_series'
life_var    = 'life_expectancy_years'
chld_var    = 'child_mortality_0_5_year_olds_dying_per_1000_born'

gdp_geo           = pd.merge(gdp,geoMap,how='left',left_on="geo",right_on="country").drop(columns='country')
gdp_geo_grp = gdp_geo.where((gdp_geo["time"]<2019) & (gdp_geo["time"]>=1800)).groupby(['world_4region','time']).agg({gdp_var :'mean'}).reset_index()


life_geo      = pd.merge(life,geoMap,how='left',left_on="geo",right_on="country").drop(columns='country')
life_geo_grp = life_geo.where((life_geo["time"]<2019) & (life_geo["time"]>=1800)).groupby(['world_4region','time']).agg({life_var :'mean'}).reset_index()

chld_mor_geo     = pd.merge(chld_mor,geoMap,how='left',left_on="geo",right_on="country").drop(columns='country')
chld_mor_geo_grp = chld_mor_geo.where((chld_mor_geo["time"]<2019) & (chld_mor_geo["time"]>=1800)).groupby(['world_4region','time']).agg({chld_var :'mean'}).reset_index()


gdp_life = pd.merge(gdp_geo_grp,life_geo_grp,
                     how='left',
                     left_on=["time","world_4region"],
                     right_on=["time","world_4region"])

gdp_life_child = pd.merge(gdp_life,chld_mor_geo_grp,
                          how='left',
                         left_on=["time","world_4region"],
                         right_on=["time","world_4region"]).rename(columns={gdp_var: "gdp", life_var: "life", chld_var: "child"})







import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd


def multiplot(data,x,y,color,x_axis,y_axis,color_code,showlegend=False,x_label='',y_label=''):
    fig = go.Figure()
    import numpy as np

    items =np.sort(list(set(data[color])))
    
    for i in range(len(items)):
        sub_data = data[data[color]==items[i]]
        fig.add_trace(go.Scatter(x=sub_data[x],
                                 y=sub_data[y],
                                 mode='lines',
                                 name=items[i],
                                 showlegend=showlegend,
                                 xaxis='x'+str(x_axis),
                                 yaxis='y'+str(y_axis),
                                 line=dict(color=color_code)))
        # fig.update_layout(
        #     xaxis=go.layout.XAxis(title=go.layout.xaxis.Title(text=x_label)),
        #     yaxis=go.layout.YAxis(title=go.layout.yaxis.Title(text=y_label)))
    return fig.data




# Initialize figure

fig = make_subplots(
        rows=2, cols=2, subplot_titles=("GDP per capita", "Child Mortality", "Life Expectancy")
    )

# Add Traces

fig.add_trace(multiplot(gdp_life_child,"time",'gdp','world_4region',1,1,"#9467bd",True)[0])
fig.add_trace(multiplot(gdp_life_child,"time",'gdp','world_4region',1,1,"#e377c2",True)[1])
fig.add_trace(multiplot(gdp_life_child,"time",'gdp','world_4region',1,1,"#bcbd22",True)[2])
fig.add_trace(multiplot(gdp_life_child,"time",'gdp','world_4region',1,1,"#17becf",True)[3])


fig.add_trace(multiplot(gdp_life_child,"time",'child','world_4region',2,2,"#9467bd")[0])
fig.add_trace(multiplot(gdp_life_child,"time",'child','world_4region',2,2,"#e377c2")[1])
fig.add_trace(multiplot(gdp_life_child,"time",'child','world_4region',2,2,"#bcbd22")[2])
fig.add_trace(multiplot(gdp_life_child,"time",'child','world_4region',2,2,"#17becf")[3])


fig.add_trace(multiplot(gdp_life_child,"time",'life','world_4region',3,3,"#9467bd")[0])
fig.add_trace(multiplot(gdp_life_child,"time",'life','world_4region',3,3,"#e377c2")[1])
fig.add_trace(multiplot(gdp_life_child,"time",'life','world_4region',3,3,"#bcbd22")[2])
fig.add_trace(multiplot(gdp_life_child,"time",'life','world_4region',3,3,"#17becf")[3])

fig.update_xaxes(title_text="years", row=1, col=1)
fig.update_xaxes(title_text="years", row=1, col=2)
fig.update_xaxes(title_text="years", row=2, col=1)

fig.update_yaxes(title_text="GDP per capita($)", row=1, col=1)
fig.update_yaxes(title_text="No. of deaths in 1000", row=1, col=2)
fig.update_yaxes(title_text="Mortality in years", row=2, col=1)


fig.update_layout(autosize=False,
                  width=1500,
                  height=900,
                  updatemenus=[go.layout.Updatemenu(
            active=0,
            buttons=list([
                dict(label="All",
                     method="update",
                     args=[{"visible": [True, True, True, True,True, True, True, True,True, True, True, True]},
                           {"title": "GDP vs Child Mortality vs Life Expectancy"}]),
                dict(label="Africa",
                     method="update",
                     args=[{"visible": [True, False, False, False,True, False, False, False,True, False, False, False]},
                           {"title": "GDP vs Child Mortality vs Life Expectancy - Africa"}]),
                dict(label="Americas",
                     method="update",
                     args=[{"visible": [False, True, False, False,False, True, False, False,False, True, False, False]},
                           {"title": "GDP vs Child Mortality vs Life Expectancy - Americas"}]),
                dict(label="Asia",
                     method="update",
                     args=[{"visible": [False, False, True, False,False, False, True, False,False, False, True, False]},
                           {"title": "GDP vs Child Mortality vs Life Expectancy - Asia"}]),
                dict(label="Europe",
                     method="update",
                     args=[{"visible": [False, False, False, True,False, False, False, True,False, False, False, True]},
                           {"title": "GDP vs Child Mortality vs Life Expectancy - Europe"}]),
            ]),
        )
    ])

# Set title
fig.update_layout(title_text="GDP vs Child Mortality vs Life Expectancy")

fig.show()
```

```python
#@title Code for 3way comparision
gdp         = pd.read_csv("ddf--datapoints--income_per_person_long_series--by--geo--time.csv").rename(columns={'income_per_person_long_series': "gdp"})
kids        = pd.read_csv("ddf--datapoints--children_per_woman_total_fertility_with_projections--by--geo--time.csv").rename(columns={'children_per_woman_total_fertility_with_projections': "kids"})
pop         = pd.read_csv("ddf--datapoints--population_total--by--geo--time.csv").rename(columns={'population_total': "population"})
geoMap      = pd.read_csv("ddf--entities--geo--country.csv")[['country','world_4region']].rename(columns={'world_4region': "continents"})

gdp_kids    = pd.merge(gdp,     kids,how='left',left_on=["geo",'time'],right_on=["geo","time"])
gdp_kids_pop= pd.merge(gdp_kids,pop ,how='left',left_on=["geo",'time'],right_on=["geo","time"])

gdp_kids_pop_geo = pd.merge(gdp_kids_pop, geoMap ,how='left',left_on=["geo"],right_on=["country"]).drop(columns=['country'])
gdp_kids_pop_geo = gdp_kids_pop_geo.dropna()
gdp_kids_pop_geo2 = gdp_kids_pop_geo[gdp_kids_pop_geo.time % 4 == 0]

# fig = px.scatter(gdp_kids_pop_geo2, x="gdp", y="kids", animation_frame="time", animation_group="geo",
#            size="population", color="continents", hover_name="geo",
#            log_x=True, size_max=55, range_x=[500,100000], range_y=[0,8])


# fig.update_layout(autosize =False,width=1300,title=go.layout.Title(text="Fertility vs GDP vs Population"),
#     xaxis=go.layout.XAxis(title=go.layout.xaxis.Title(text="GDP per capita")),
#     yaxis=go.layout.YAxis(title=go.layout.yaxis.Title(text="Fertility")))

# fig.show()

fig = px.scatter(gdp_kids_pop_geo2, x="gdp", y="kids", animation_frame="time", animation_group="geo",
           size="population", color="continents", hover_name="geo",
           log_x=True, size_max=55, range_x=[500,100000], range_y=[0,8])

fig.update_layout(title_text='Fertility vs GDP vs Population',
                  xaxis=dict(title='GDP per Capita'),
                  yaxis=dict(title='Children per woman'))

fig
```
